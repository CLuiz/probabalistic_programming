```python
from matplotlib import pyplot as plt
from IPython.core.pylabtools import figsize
import numpy as np
%matplotlib inline
```

```python
figsize(12.5, 4)
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.dpi'] = 300
colors = ["#348ABD", "#A60628"]
```

```python
prior = [1/21., 20/21.]
posterior = [0.087, 1 - 0.087]
plt.bar([0, .7], prior, alpha=.70, width=0.25,
        color=colors[0], label='prior distribution',
        lw="3", edgecolor=colors[0])

plt.bar([.25, .95], posterior, alpha=.70, width=0.25,
        color=colors[1], label='posterior distribution',
        lw="3", edgecolor=colors[1])
plt.xticks([.2, .95], ["Librarian", "Farmer"])
plt.title("Prior and posterior probabilities of Steve's occuptation")
plt.ylabel("Probability")
plt.legend(loc="upper left");
```
